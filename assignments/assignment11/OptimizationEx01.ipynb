## Hat potential

The following potential is often used in Physics and other fields to describe symmetry breaking and is often known as the "hat potential":

$$ V(x) = -a x^2 + b x^4 $$

Write a function `hat(x,a,b)` that returns the value of this function:

```python
def hat(x,a,b):
    return -a*x**2 + b*x**4
```

```python
assert hat(0.0, 1.0, 1.0)==0.0
assert hat(0.0, 1.0, 1.0)==0.0
assert hat(1.0, 10.0, 1.0)==-9.0
```

Plot this function over the range $x\in\left[-3,3\right]$ with $b=1.0$ and $a=5.0$:

```python
a = 5.0
b = 1.0
```

```python
x = np.linspace(-3,3,100)
y = hat(x,5.0,1.0)
def plot(x,y):
    plt.figure(figsize=(10.0,8.0))
    plt.title("Hat Potential")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.xlim(-4,4)
    plt.ylim(-7,10)
    plt.box(False)
    plt.grid(True)
    plt.plot(x,y)
plot(x,y)
```

```python
x1 = opt.minimize_scalar(hat,bracket=[-2,0],args=(5.0,1.0))
x2 = opt.minimize_scalar(hat,bracket=[0,2],args=(5.0,1.0))
x1min,x2min = x1.x,x2.x
y1,y2 = hat(x1min,5.0,1.0),hat(x2min,5.0,1.0)
print("The minimum values of x are %f and %f" % (x1min,x2min))
plot(x,y)
plt.scatter(x1min,y1,marker="o",color="r")
plt.scatter(x2min,y2,marker="o",color="r")
```

$$ V(x) = -a x^2 + b x^4 $$

To find the results analytically take the derivative and set it equal to zero.

$$ \frac{dV(x)}{dx} = -2ax + 4bx^3 = 0 $$

One solution is x = 0, which we discard because this corresponds to a maximum, not a minimum.
Thus we can divide by x, requiring that the following solutions are not equal to zero.

$$ -2a + 4bx^2 = 0 $$ 
$$ a = 2bx^2 $$
$$ x = \pm\sqrt{\frac{a}{2b}} $$
